````python
#import all necessary labraries
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np
````

## **Scrapping Movies Data**

````python
# Specifying the URL from which movies related data will be fetched
url='https://www.justwatch.com/in/movies?release_year_from=2000'

# Sending an HTTP GET request to the URL
page=requests.get(url)
# Parsing the HTML content using BeautifulSoup with the 'html.parser'
soup=BeautifulSoup(page.text,'html.parser')
# Printing the prettified HTML content
# print(soup.prettify())
````

## **Fetching Movie URL's**

````python
# Write Your Code here
a = soup.find_all('a', class_ = 'title-list-grid__item--link')

movie_url = []
for i in range(0,len(a)):
  movie_url.append(f"https://www.justwatch.com{a[i].get('href')}")

print(movie_url)
````

## **Scrapping Movie Title**

````python
# Write Your Code here
title_head = soup.find_all('div', class_ = 'title-list-grid__item')

movie_title = []
for i in range(0,len(a)):
  movie_title.append(f"{title_head[i].get('data-title')}")

print(movie_title)
````

## **Scrapping release Year**

````python
# Write Your Code here
release_year = []
for i in range(0,len(movie_url)):
  # Sending an HTTP GET request to the URL
  page_ry=requests.get(movie_url[i])
  # Parsing the HTML content using BeautifulSoup with the 'html.parser'
  soup_ry=BeautifulSoup(page_ry.text, 'html.parser')
  # print(soup_ry.text)
  h1_ry = soup_ry.find('span', class_="text-muted")
  # print(h1_ry)
  if h1_ry is None:
    span_ry = 'NA'
  else:
    span_ry = h1_ry.text
  release_year.append(span_ry.split("(")[-1].split(")")[0])
print(release_year)
````

## **Scrapping Genres**

````python
# Write Your Code here
Genres = ['NA'] * len(movie_url)
for i in range(0,len(movie_url)):
  # Sending an HTTP GET request to the URL
  page_Genres = requests.get(movie_url[i])
  # Parsing the HTML content using BeautifulSoup with the 'html.parser'
  soup_Genres = BeautifulSoup(page_Genres.text, 'html.parser')
  # Finding all div elements with class 'detail-infos'
  Genres_tag = soup_Genres.find_all('h3', class_='detail-infos__subheading')
  # print(Genres_tag)

  # print(Genres_tag)
  for Genre in Genres_tag:
    # print(Genre.text)
    if Genres_tag and 'Genres' in Genre.text:  # means inside subheading class and in that tag if Genres text is present
      # Finding the values of Genres
      Genres_div = Genre.find_next_sibling('div')
      Genres_values = Genres_div.find('span', class_ = "detail-infos__value")
      # print(Genres_values)

      # Storing in the list
      if Genres_values:
        Genres[i] = Genres_values.text.strip()
      # print(Genres_values)

print(Genres)
````

## **Scrapping IMBD Rating**

````python
# Write Your Code here
IMDB_rating = ['NA']* len(movie_url)
for i in range(0,len(movie_url)):
  # Sending an HTTP GET request to the URL
  page_imdb = requests.get(movie_url[i])
  # Parsing the HTML content using BeautifulSoup with the 'html.parser'
  soup_imdb = BeautifulSoup(page_imdb.text, 'html.parser')
  # print(soup_imdb)
  imdb_tag = soup_imdb.find('h3',class_="detail-infos__subheading", string = 'Rating')
  # print(imdb_tag)
  if imdb_tag:
  # finding a imdb values from text in sub-tag and then trimming extra spaces from ends
    imdb_values = imdb_tag.find_next_sibling('div', class_="detail-infos__value").text.strip()
    imdb_value = imdb_values.split( )
    # print(imdb_value)
    IMDB_rating[i] = float(imdb_value[0])

# IMDB_Rating = dict(zip(movie_url,IMDB_rating))
print(IMDB_rating)
````

## **Scrapping Runtime/Duration**

````python
# Write Your Code here
Runtime = []
for i in range(0,len(movie_url)):
  # Sending an HTTP GET request to the URL
  page_run = requests.get(movie_url[i])
  # Parsing the HTML content using BeautifulSoup with the 'html.parser'
  soup_run = BeautifulSoup(page_run.text, 'html.parser')
  # print(soup_run)
  run_tag = soup_run.find('h3', class_="detail-infos__subheading", string = 'Runtime')
  # print(run_tag)
  if run_tag is None:
     run_values = 'NA'
  else:
  # finding a run values from text in sub-tag and then trimming extra spaces from ends
    run_values = run_tag.find_next_sibling('div', class_="detail-infos__value").text.strip()
  Runtime.append(run_values)

print(Runtime)
````

## **Scrapping Age Rating**

````python
# Write Your Code here
Age_Rating = []
for i in range(0,len(movie_url)):
  # Sending an HTTP GET request to the URL
  page_age_rating = requests.get(movie_url[i])
  # Parsing the HTML content using BeautifulSoup with the 'html.parser'
  soup_age_rating = BeautifulSoup(page_age_rating.text, 'html.parser')
  # print(soup_age_rating)
  age_rating_tag = soup_age_rating.find('h3', class_="detail-infos__subheading", string = 'Age rating')
  # print(age_rating_tag)
  if age_rating_tag is None:
     age_rating_values = 'NA'
  else:
  # finding a age_rating values from text in sub-tag and then trimming extra spaces from ends
    age_rating_values = age_rating_tag.find_next_sibling('div', class_="detail-infos__value").text.strip()
  Age_Rating.append(age_rating_values)

print(Age_Rating)
````

## **Fetching Production Countries Details**

## **Fetching Streaming Service Details**

````python
# Write Your Code here
Streaming = ['NA'] * len(movie_url)
for i in range(0,len(movie_url)):
  # Sending an HTTP GET request to the URL
  page_streaming = requests.get(movie_url[i])
  # Parsing the HTML content using BeautifulSoup with the 'html.parser'
  soup_streaming = BeautifulSoup(page_streaming.text, 'html.parser')
  # Finding all div elements with class 'detail-infos'
  streaming_tag = soup_streaming.find_all('img', class_="offer__icon")
  # print(streaming_tag)
  streaming = []
  for stream in streaming_tag:
     # Finding the tag which contains subheadings because Age rating is contained with subheading class
    stream_sub = stream.get('alt')
    #  print(stream_sub)
    if stream_sub != 'Bookmyshow' and stream_sub != 'Amazon Video':
      streaming.append(stream_sub)
    if streaming:
      Streaming[i] = ', '.join(list(set(streaming)))

print(Streaming)
# print(len(Streaming))
````

## **Analyzing Top Genres**

````python
# Write Your Code here
df_cleaned_genre = df[df['Genres'] != 'NA']
Genres_column = df_cleaned_genre['Genres']
# Genres_column
Genre_list = []
for genre_element in Genres_column:
  genre_list = genre_element.split(", ")
  Genre_list += genre_list

from collections import Counter
# Use Counter to count occurrences
Genre_element_counts = Counter(Genre_list)
# Genre_element_counts
# Sort by values in descending order using most_common()
sorted_elements = Genre_element_counts.most_common()

for element, count in sorted_elements:
    print(f"{element}: {count}")
````
